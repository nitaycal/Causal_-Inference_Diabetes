### Creating Results File

```python
import pandas as pd
```

```python
results = pd.read_csv('models_results.csv')
results
```

```python
df = results.groupby(['train', 'default', 'method']).agg({'readmitted':'mean', 'model_acc':'mean', 'ate0':'mean', 'ate1':'mean', 'ate':'mean'}).reset_index()
df['model'] = 'LogisticRegression'
df.loc[df['default'] == False ,'model'] = 'RandomForests'
df.loc[df['method'].str.contains('matching') ,'model'] = 'KNN'
df = df.rename(columns={'ate0':'ATE0', 'ate1':'ATE1', 'ate':'ATE'})
df = df.sort_values(by=['method', 'train', 'model'])[['train', 'model', 'method', 'model_acc', 'readmitted', 'ATE0', 'ATE1', 'ATE']]
```

```python
df.to_csv('results.csv', index=False)
```
